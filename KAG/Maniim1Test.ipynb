```python
# Testing simple animations
from manim import *
```

```python

%%manim -qh -v WARNING ListComprehensionAnimation

class ListComprehensionAnimation(Scene):
    def construct(self):
        # Title
        title = Text("List Comprehension in Python", font_size=36)
        title.to_edge(UP)
        self.play(Write(title))
        self.wait(1)

        # List comprehension explanation
        explanation = Text("List comprehension allows us to create a list based on existing lists.", font_size=24)
        explanation.next_to(title, DOWN)
        self.play(Write(explanation))
        self.wait(2)

        # Original list
        original_list = Text("Original List: [1, 2, 3, 4, 5]", font_size=24)
        original_list.next_to(explanation, DOWN)
        self.play(Write(original_list))
        self.wait(1)

        # List comprehension code
        code = Text("Squared List: [x**2 for x in Original List]", font_size=24)
        code.next_to(original_list, DOWN)
        self.play(Write(code))
        self.wait(1)

        # Result list
        result_list = Text("Result List: [1, 4, 9, 16, 25]", font_size=24)
        result_list.next_to(code, DOWN)
        self.play(Transform(code, result_list))
        self.wait(2)

        # Fade out all objects
        self.play(FadeOut(title), FadeOut(explanation), FadeOut(original_list), FadeOut(code))

# To render the animation with high quality and 60 FPS, you can use the following command in your terminal:
# manim -pqh your_script_name.py ListComprehensionAnimation
```
